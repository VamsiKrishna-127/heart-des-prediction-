```python
from tkinter import *
import joblib
```

```python
import joblib
from tkinter import Tk, Label, Entry, Button

def validate_input(entry):
    """Helper function to check if the input is a valid integer."""
    try:
        return int(entry.get())
    except ValueError:
        return None

def show_entry_fields():
    try:
        # Validate each input and collect values
        age = validate_input(age_entry)
        gender = validate_input(gender_entry)
        chest_pain_type = validate_input(chest_pain_entry)
        resting_blood_pressure = validate_input(bp_entry)
        cholesterol = validate_input(cholesterol_entry)
        fasting_blood_sugar = validate_input(fbs_entry)
        rest_ecg = validate_input(restecg_entry)
        max_heart_rate = validate_input(max_hr_entry)
        exercise_induced_angina = validate_input(exang_entry)
        old_peak = validate_input(oldpeak_entry)
        slope = validate_input(slope_entry)
        major_vessels = validate_input(vessels_entry)
        thalassemia = validate_input(thal_entry)
        
        # Add the missing features here 
        feature_14 = validate_input(feature14_entry)
        feature_15 = validate_input(feature15_entry)

        # If any feature is invalid, show an error and return
        if None in [age, gender, chest_pain_type, resting_blood_pressure, cholesterol,
                    fasting_blood_sugar, rest_ecg, max_heart_rate, exercise_induced_angina,
                    old_peak, slope, major_vessels, thalassemia, feature_14, feature_15]:
            Label(master, text="Error: Please enter valid numeric values for all fields", fg="red").grid(row=31)
            return

        # Load the model
        model = joblib.load('model_joblib_heart')

        # Predict using the model
        result = model.predict([[
            age, gender, chest_pain_type, resting_blood_pressure, cholesterol,
            fasting_blood_sugar, rest_ecg, max_heart_rate, exercise_induced_angina,
            old_peak, slope, major_vessels, thalassemia, feature_14, feature_15
        ]])

        # Display the result
        if result[0] == 0:
            Label(master, text="No Heart Disease Detected", fg="green").grid(row=31)
        else:
            Label(master, text="High Risk of Heart Disease", fg="red").grid(row=31)
    
    except Exception as e:
        Label(master, text=f"Error: {str(e)}", fg="red").grid(row=31)

# Create main window
master = Tk()
master.title("Heart Disease Prediction System")

# Create labels and entry fields
Label(master, text="Heart Disease Prediction System", bg="black", fg="white").grid(row=0, columnspan=2)
Label(master, text="Enter Your Age").grid(row=1)
Label(master, text="Enter Gender (1 for Male, 0 for Female)").grid(row=2)
Label(master, text="Enter Chest Pain Type (0-3)").grid(row=3)
Label(master, text="Enter Resting Blood Pressure (mm Hg)").grid(row=4)
Label(master, text="Enter Cholesterol Level (mg/dL)").grid(row=5)
Label(master, text="Enter Fasting Blood Sugar (1 if > 120 mg/dL, 0 otherwise)").grid(row=6)
Label(master, text="Enter Resting ECG Results (0-2)").grid(row=7)
Label(master, text="Enter Maximum Heart Rate Achieved").grid(row=8)
Label(master, text="Enter Exercise-Induced Angina (1 for Yes, 0 for No)").grid(row=9)
Label(master, text="Enter Old Peak (ST depression induced by exercise)").grid(row=10)
Label(master, text="Enter Slope of the Peak Exercise ST Segment (0-2)").grid(row=11)
Label(master, text="Enter Number of Major Vessels Colored by Fluoroscopy (0-3)").grid(row=12)
Label(master, text="Enter Thalassemia (3 = Normal, 6 = Fixed Defect, 7 = Reversible Defect)").grid(row=13)

# Create entry fields
age_entry = Entry(master)
gender_entry = Entry(master)
chest_pain_entry = Entry(master)
bp_entry = Entry(master)
cholesterol_entry = Entry(master)
fbs_entry = Entry(master)
restecg_entry = Entry(master)
max_hr_entry = Entry(master)
exang_entry = Entry(master)
oldpeak_entry = Entry(master)
slope_entry = Entry(master)
vessels_entry = Entry(master)
thal_entry = Entry(master)

# Create new input fields for feature_14 and feature_15
feature14_entry = Entry(master)
feature15_entry = Entry(master)

# Place entry fields on the grid
age_entry.grid(row=1, column=1)
gender_entry.grid(row=2, column=1)
chest_pain_entry.grid(row=3, column=1)
bp_entry.grid(row=4, column=1)
cholesterol_entry.grid(row=5, column=1)
fbs_entry.grid(row=6, column=1)
restecg_entry.grid(row=7, column=1)
max_hr_entry.grid(row=8, column=1)
exang_entry.grid(row=9, column=1)
oldpeak_entry.grid(row=10, column=1)
slope_entry.grid(row=11, column=1)
vessels_entry.grid(row=12, column=1)
thal_entry.grid(row=13, column=1)

# Place the new input fields for feature_14 and feature_15
feature14_entry.grid(row=14, column=1)
feature15_entry.grid(row=15, column=1)

# Create predict button
Button(master, text='Predict', command=show_entry_fields).grid(row=30, column=1)

# Start the GUI loop
master.mainloop()
```
